```python
## Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from scipy import stats
from numpy.core.fromnumeric import mean
```

```python
df_medical = pd.read_csv("Data/Data_Medial.csv")
df_medical.head()
```

```python
## Creating first graph to explore VitD_levels feature

## Calculating Stats
col = "VitD_levels"
feature = df_medical[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
std_plus = mean + std
std_minus = mean - std 

## Creating Graph
fig, ax = plt.subplots(figsize = (20, 10))
ax = sns.histplot(feature, ax = ax, stat = "probability", kde = True) 
ax.axvline(mean, color = "red", label = f'Mean = {mean}');
ax.axvline(median, color = "green", label = f'Median = {median}')
ax.axvline(std_plus, color = "black", label = f'std + 1 = {std_plus}');
ax.axvline(std_minus, color = "black", label = f'std - 1 = {std_minus}')
ax.axvspan(std_plus, std_minus, color = "yellow", zorder = 0)
ax.set_title(f'{col}')
ax.legend();
```

```python
## Finding kurtosis
kurt = stats.kurtosis(df_medical["VitD_levels"], fisher = False)
kurt
```

## Graph One Analysis:
---
1. Continuous
2. Normal - no skew 
3. Mesokurtic with a score of 3.013
